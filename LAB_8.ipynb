4. Viết chương trình Python với thư viện NLTK để loại bỏ các stopword từ một văn bản đã cho.

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')     # Nếu Colab yêu cầu
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Văn bản cần xử lý
text = """
Natural Language Processing is one of the most exciting fields of Artificial Intelligence.
It helps computers understand human language.
"""

# Chọn ngôn ngữ
language = "english"

# Tách từ
words = word_tokenize(text)

# Danh sách stopwords
stop_words = set(stopwords.words(language))

# Loại bỏ stopwords
filtered_words = []

for word in words:
    if word.lower() not in stop_words:
        filtered_words.append(word)

print("Văn bản gốc:\n")
print(text)

print("\nCác từ sau khi loại bỏ stopwords:\n")
print(filtered_words)

print("\nVăn bản sau khi loại bỏ stopwords:\n")
print(" ".join(filtered_words))
```

5. Viết chương trình Python với thư viện NLTK bỏ qua các stopword từ danh sách các stopword.

```python
from nltk.corpus import stopwords

# Lấy danh sách stopwords tiếng Anh
stop_words = set(stopwords.words('english'))

print("Số lượng stopwords ban đầu:", len(stop_words))

# Các stopword muốn bỏ qua (không loại bỏ khỏi văn bản)
remove_words = {"no", "not", "nor"}

# Tạo danh sách stopwords mới
new_stopwords = stop_words - remove_words

print("\nCác từ được giữ lại:")
print(remove_words)

print("\nSố lượng stopwords sau khi cập nhật:", len(new_stopwords))

print("\nKiểm tra:")
print("'not' có trong stopwords mới không?",
      "not" in new_stopwords)

print("'the' có trong stopwords mới không?",
      "the" in new_stopwords)
```

6. Viết một chương trình Python với thư viện NLTK để tìm định nghĩa và ví dụ của một từ đã cho bằng WordNet từ Wikipedia

```python
import nltk

nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.corpus import wordnet as wn

# Nhập từ cần tra
word = input("Nhập một từ tiếng Anh: ")

# Tìm các synset của từ
synsets = wn.synsets(word)

if len(synsets) == 0:
    print(f"Không tìm thấy từ '{word}' trong WordNet.")
else:
    print(f"\nCó {len(synsets)} nghĩa của từ '{word}':\n")

    for i, syn in enumerate(synsets, start=1):

        print("=" * 60)
        print(f"Nghĩa {i}")

        print("Synset      :", syn.name())
        print("Định nghĩa  :", syn.definition())

        examples = syn.examples()

        if examples:
            print("Ví dụ       :", examples[0])
        else:
            print("Ví dụ       : Không có")
```

7. Viết chương trình Python với thư viện NLTK để tìm tập hợp các từ đồng nghĩa và trái nghĩa của một từ nào đó.

```python
from nltk.corpus import wordnet as wn

# Nhập từ cần tìm
word = input("Nhập một từ tiếng Anh: ")

# Tập hợp lưu từ đồng nghĩa và trái nghĩa
synonyms = set()
antonyms = set()

# Duyệt tất cả các synset của từ
for syn in wn.synsets(word):

    # Duyệt từng lemma trong synset
    for lemma in syn.lemmas():

        # Thêm từ đồng nghĩa
        synonyms.add(lemma.name())

        # Nếu có từ trái nghĩa thì thêm vào
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

# Hiển thị kết quả
print("\n========== KẾT QUẢ ==========")

print("\nTừ cần tìm:", word)

print("\nCác từ đồng nghĩa:")
if synonyms:
    for s in sorted(synonyms):
        print("-", s)
else:
    print("Không có.")

print("\nCác từ trái nghĩa:")
if antonyms:
    for a in sorted(antonyms):
        print("-", a)
else:
    print("Không có.")
```

9. Viết chương trình Python với thư viện NLTK để so sánh sự giống nhau của hai danh từ đã cho.

```python
from nltk.corpus import wordnet as wn

# Nhập hai danh từ
noun1 = input("Nhập danh từ thứ nhất: ")
noun2 = input("Nhập danh từ thứ hai: ")

# Lấy nghĩa đầu tiên của mỗi danh từ
synsets1 = wn.synsets(noun1, pos=wn.NOUN)
synsets2 = wn.synsets(noun2, pos=wn.NOUN)

if not synsets1:
    print(f"Không tìm thấy danh từ '{noun1}' trong WordNet.")
elif not synsets2:
    print(f"Không tìm thấy danh từ '{noun2}' trong WordNet.")
else:
    syn1 = synsets1[0]
    syn2 = synsets2[0]

    similarity = syn1.wup_similarity(syn2)

    print("\n========== KẾT QUẢ ==========")
    print("Danh từ 1 :", noun1)
    print("Danh từ 2 :", noun2)
    print("Synset 1  :", syn1.name())
    print("Synset 2  :", syn2.name())

    if similarity is not None:
        print(f"\nĐộ giống nhau (Wu-Palmer): {similarity:.4f}")
    else:
        print("\nKhông thể tính độ giống nhau.")
```

10. Viết chương trình Python với thư viện NLTK để so sánh sự giống nhau của hai động từ đã cho.

```python
from nltk.corpus import wordnet as wn

# Nhập hai động từ
verb1 = input("Nhập động từ thứ nhất: ")
verb2 = input("Nhập động từ thứ hai: ")

# Lấy các nghĩa của hai động từ
synsets1 = wn.synsets(verb1, pos=wn.VERB)
synsets2 = wn.synsets(verb2, pos=wn.VERB)

if not synsets1:
    print(f"Không tìm thấy động từ '{verb1}' trong WordNet.")
elif not synsets2:
    print(f"Không tìm thấy động từ '{verb2}' trong WordNet.")
else:
    syn1 = synsets1[0]
    syn2 = synsets2[0]

    similarity = syn1.wup_similarity(syn2)

    print("\n========== KẾT QUẢ ==========")
    print("Động từ 1 :", verb1)
    print("Động từ 2 :", verb2)
    print("Synset 1  :", syn1.name())
    print("Synset 2  :", syn2.name())

    if similarity is not None:
        print(f"\nĐộ giống nhau (Wu-Palmer): {similarity:.4f}")
    else:
        print("\nKhông thể tính độ giống nhau.")
```

12. Viết chương trình Python với thư viện NLTK để in 15 kết hợp ngẫu nhiên đầu tiên được gắn nhãn nam và được gắn nhãn tên nữ từ kho tên.

```python
from nltk.corpus import names
import random

# Tạo danh sách (tên, nhãn)
labeled_names = ([(name, 'male') for name in names.words('male.txt')] +
                 [(name, 'female') for name in names.words('female.txt')])

# Trộn ngẫu nhiên dữ liệu
random.shuffle(labeled_names)

# In 15 phần tử đầu tiên
print("15 kết hợp ngẫu nhiên đầu tiên:\n")

for i, (name, gender) in enumerate(labeled_names[:15], start=1):
    print(f"{i}. ({name}, {gender})")
```

13. Viết chương trình Python với thư viện NLTK để trích xuất ký tự cuối cùng của tất cả các tên được gắn nhãn và tạo mảng mới với chữ cái cuối cùng của mỗi tên và nhãn được liên kết.

```python
from nltk.corpus import names

# Tạo danh sách (tên, nhãn)
labeled_names = ([(name, 'male') for name in names.words('male.txt')] +
                 [(name, 'female') for name in names.words('female.txt')])

# Tạo danh sách mới gồm (ký tự cuối, nhãn)
last_letter_data = []

for name, gender in labeled_names:
    last_letter_data.append((name[-1], gender))

# Hiển thị kết quả
print("10 phần tử đầu tiên của mảng mới:\n")

for i, item in enumerate(last_letter_data[:10], start=1):
    print(f"{i}. {item}")

print("\nTổng số phần tử:", len(last_letter_data))
```
