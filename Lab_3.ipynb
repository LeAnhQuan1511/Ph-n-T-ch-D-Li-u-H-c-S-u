# LAB 3: LÀM SẠCH DỮ LIỆU CƠ BẢN

**Vấn đề 1: Tải dữ liệu và xử lý thiếu tiêu đề (Missing header)**


```python
import pandas as pd

# Định nghĩa tên các cột cho dữ liệu
column_names = ['ID', 'Name', 'Age', 'Weight', 'm0006', 'm0612', 'm1218', 'f0006', 'f0612', 'f1218']

# Tải dữ liệu từ file csv, xử lý missing header và bỏ qua các dòng lỗi (bad lines)
df = pd.read_csv('/content/patient_heart_rate.csv', names=column_names, on_bad_lines='warn')

# Hiển thị 5 dòng đầu tiên
display(df.head())
```

**Vấn đề 2**

```python
# Tách cột Name thành Firstname và Lastname
df[['Firstname', 'Lastname']] = df['Name'].str.split(expand=True, n=1)

# Xóa cột Name gốc
df = df.drop('Name', axis=1)

# Hiển thị kết quả
display(df.head())
```

**Vấn đề 3**

```python
def convert_weight(weight):
    if pd.isna(weight):
        return weight

    weight_str = str(weight).lower()
    if 'kgs' in weight_str:
        return float(weight_str.replace('kgs', ''))
    elif 'lbs' in weight_str:
        # Chuyển đổi lbs sang kg: 1 lb = 0.453592 kg
        return float(weight_str.replace('lbs', '')) * 0.453592
    return float(weight_str)

# Áp dụng hàm chuyển đổi và cập nhật đơn vị
df['Weight'] = df['Weight'].apply(convert_weight)

# Hiển thị kết quả
display(df.head())
```

**Vấn đề 4**

```python
# Xóa bỏ các dòng dữ liệu có chứa giá trị NaN
df.dropna(how='all', inplace=True)

# Hiển thị kết quả sau khi xóa
display(df)
```

**Vấn đề 5**

```python
# Xử lý dữ liệu trùng lặp hoàn toàn
df.drop_duplicates(inplace=True)

# Hiển thị kết quả sau khi loại bỏ trùng lặp
display(df)
```

**Vấn đề 6**

```python
# Xử lý ký tự non-ASCII trong cột Firstname và Lastname
df['Firstname'] = df['Firstname'].replace(r'[^\x00-\x7F]+', '', regex=True)
df['Lastname'] = df['Lastname'].replace(r'[^\x00-\x7F]+', '', regex=True)

# Hiển thị kết quả
display(df.head())
```

**Vấn đề 7: Xử lý Missing Values trên Age và Weight**

```python
# 1. Thống kê dữ liệu thiếu
print("Thống kê dữ liệu thiếu:")
print(df[['Age', 'Weight']].isna().sum())

# 2. Xóa dòng nếu thiếu cả Age và Weight
df.dropna(subset=['Age', 'Weight'], how='all', inplace=True)

# 3. Thay thế Age thiếu bằng giá trị mean
# Chuyển đổi Age sang numeric nếu cần thiết trước khi tính mean
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
age_mean = df['Age'].mean()
df['Age'].fillna(age_mean, inplace=True)

# Hiển thị kết quả
display(df.head())
print(f"\nGiá trị Age dùng để thay thế (Mean): {age_mean}")
```

**Vấn đề 8**

```python
# Chuyển dữ liệu từ wide format sang long format

df = pd.melt(
    df,
    id_vars=["Firstname", "Lastname", "Age", "Weight"],
    value_vars=["m0006","m0612","m1218",
                "f0006","f0612","f1218"],
    var_name="Code",
    value_name="PulseRate"
)

# Tách giới tính
df["Sex"] = df["Code"].str[0]

# Tách thời gian
df["Time"] = df["Code"].str[1:]

# Đổi giới tính
df["Sex"] = df["Sex"].replace({
    "m":"Male",
    "f":"Female"
})

# Đổi mã thời gian
df["Time"] = df["Time"].replace({
    "0006":"00-06",
    "0612":"06-12",
    "1218":"12-18"
})

# Loại bỏ giá trị không hợp lệ
df["PulseRate"] = df["PulseRate"].replace("-", pd.NA)

# Chuyển PulseRate sang số
df["PulseRate"] = pd.to_numeric(
    df["PulseRate"],
    errors="coerce"
)

# Xóa các dòng không có PulseRate
df = df.dropna(subset=["PulseRate"])

# Đưa PulseRate về int
df["PulseRate"] = df["PulseRate"].astype(int)

# Loại bỏ cột Code
df = df.drop(columns=["Code"])

# Sắp xếp dữ liệu
df = df.sort_values(
    by=["Firstname","Lastname","Sex","Time"]
)

# Đánh lại chỉ số
df = df.reset_index(drop=True)

print(df)
```

11. Hãy khảo sát tỉ lệ dữ liệu thiếu trên biến huyết áp. Dữ liệu bị thiếu thì hãy xử lý bằng
phương pháp sau **bold text**

```python
df = df.sort_values(
    by=['Firstname','Lastname','Sex','Time']
).reset_index(drop=True)
```

```python
import pandas as pd
import numpy as np

for idx in df[df['PulseRate'].isna()].index:

    person = (
        (df['Firstname'] == df.loc[idx,'Firstname']) &
        (df['Lastname'] == df.loc[idx,'Lastname']) &
        (df['Sex'] == df.loc[idx,'Sex'])
    )

    person_data = df[person]

    person_idx = person_data.index.tolist()

    pos = person_idx.index(idx)

    value = np.nan

    # 1. Trung bình trước và sau
    if pos > 0 and pos < len(person_idx)-1:

        prev_val = df.loc[person_idx[pos-1],'PulseRate']
        next_val = df.loc[person_idx[pos+1],'PulseRate']

        if pd.notna(prev_val) and pd.notna(next_val):
            value = (prev_val + next_val)/2

    # 2. Trung bình 2 giá trị trước
    if pd.isna(value):

        if pos >= 2:

            p1 = df.loc[person_idx[pos-1],'PulseRate']
            p2 = df.loc[person_idx[pos-2],'PulseRate']

            if pd.notna(p1) and pd.notna(p2):
                value = (p1+p2)/2

    # 3. Trung bình 2 giá trị sau
    if pd.isna(value):

        if pos <= len(person_idx)-3:

            n1 = df.loc[person_idx[pos+1],'PulseRate']
            n2 = df.loc[person_idx[pos+2],'PulseRate']

            if pd.notna(n1) and pd.notna(n2):
                value = (n1+n2)/2

    # 4. Trung bình của người đó
    if pd.isna(value):

        mean_person = person_data['PulseRate'].mean()

        if pd.notna(mean_person):
            value = mean_person

    # 5. Trung bình theo giới tính
    if pd.isna(value):

        mean_sex = (
            df[df['Sex']==df.loc[idx,'Sex']]
            ['PulseRate']
            .mean()
        )

        if pd.notna(mean_sex):
            value = mean_sex

    # 6. Trung bình toàn bộ dữ liệu
    if pd.isna(value):

        mean_all = df['PulseRate'].mean()

        if pd.notna(mean_all):
            value = mean_all

    # 7. Giá trị ổn định y học
    if pd.isna(value):

        value = 75

    df.loc[idx,'PulseRate'] = round(value)
```
