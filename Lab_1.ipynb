5. Thống kê dữ liệu thiếu cho biến T1 và hiệu chỉnh dữ liệu, lưu ý việc thay thế dữ liệu thiếu sử
dụng phương pháp Mean.

```python
import pandas as pd

# Đọc dữ liệu
df = pd.read_csv("dulieuxettuyendaihoc.csv")

# ===== 1. THỐNG KÊ DỮ LIỆU THIẾU =====

print("===== THỐNG KÊ DỮ LIỆU THIẾU CỘT T1 =====")

# Đếm số lượng dữ liệu thiếu
print("Số lượng dữ liệu thiếu:")
print(df["T1"].isnull().sum())

# ===== BẢNG TẦN SỐ =====
print("\n===== BẢNG TẦN SỐ =====")
print(df["T1"].value_counts(dropna=False))

# ===== BẢNG TẦN SUẤT =====
print("\n===== BẢNG TẦN SUẤT =====")
print(df["T1"].value_counts(dropna=False, normalize=True) * 100)

# ===== GIÁ TRỊ RIÊNG BIỆT =====
print("\n===== UNIQUE VALUES =====")
print(df["T1"].unique())

# ===== 2. THAY THẾ DỮ LIỆU THIẾU BẰNG MEAN =====

# Tính giá trị trung bình
mean_value = df["T1"].mean()

print("\nGiá trị trung bình của T1:")
print(mean_value)

# Thay thế dữ liệu thiếu bằng Mean
df["T1"] = df["T1"].fillna(mean_value)

# ===== KIỂM TRA SAU KHI XỬ LÝ =====
print("\n===== SAU KHI XỬ LÝ =====")
print("Số lượng dữ liệu thiếu:")
print(df["T1"].isnull().sum())

# Hiển thị 10 dòng đầu
print("\n===== 10 DÒNG ĐẦU =====")
print(df[["T1"]].head(10))
```

6. Hãy thực hiện xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại.

```python
import pandas as pd

# Đọc dữ liệu
df = pd.read_csv("dulieuxettuyendaihoc.csv")

# Danh sách các cột điểm số
score_columns = [
    'T1','L1','H1','S1','V1','X1','D1','N1',
    'T2','L2','H2','S2','V2','X2','D2','N2',
    'T3','L3','H3','S3','V3','X3','D3','N3',
    'T4','L4','H4','S4','V4','X4','D4','N4',
    'T5','L5','H5','S5','V5','X5','D5','N5',
    'T6','L6','H6','S6','V6','X6','D6','N6',
    'DH1','DH2','DH3'
]

print("===== THỐNG KÊ DỮ LIỆU THIẾU =====")

# Kiểm tra dữ liệu thiếu trước khi xử lý
for col in score_columns:
    missing = df[col].isnull().sum()
    print(f"{col}: {missing}")

# ===== XỬ LÝ DỮ LIỆU THIẾU =====

for col in score_columns:

    # Tính mean của từng cột
    mean_value = df[col].mean()

    # Thay thế giá trị thiếu bằng mean
    df[col] = df[col].fillna(mean_value)

print("\n===== SAU KHI XỬ LÝ =====")

# Kiểm tra lại
for col in score_columns:
    missing = df[col].isnull().sum()
    print(f"{col}: {missing}")

# Hiển thị 10 dòng đầu
print("\n===== 10 DÒNG ĐẦU =====")
print(df[score_columns].head(10))
```

8. Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp
10, 11, 12 như sau:

```python
import pandas as pd

# Đọc dữ liệu
df = pd.read_csv("dulieuxettuyendaihoc.csv")

# ===== TẠO TBM1 =====
df["TBM1"] = (
    df["T1"] * 2 +
    df["L1"] +
    df["H1"] +
    df["S1"] +
    df["V1"] * 2 +
    df["X1"] +
    df["D1"] +
    df["N1"]
) / 10

# ===== TẠO TBM2 =====
df["TBM2"] = (
    df["T2"] * 2 +
    df["L2"] +
    df["H2"] +
    df["S2"] +
    df["V2"] * 2 +
    df["X2"] +
    df["D2"] +
    df["N2"]
) / 10

# ===== TẠO TBM3 =====
df["TBM3"] = (
    df["T6"] * 2 +
    df["L6"] +
    df["H6"] +
    df["S6"] +
    df["V6"] * 2 +
    df["X6"] +
    df["D6"] +
    df["N6"]
) / 10

# ===== HÀM XẾP LOẠI =====
def xep_loai(tbm):

    if tbm < 5:
        return "Y"

    elif tbm < 6.5:
        return "TB"

    elif tbm < 8:
        return "K"

    elif tbm < 9:
        return "G"

    else:
        return "XS"

# ===== TẠO XL1 XL2 XL3 =====
df["XL1"] = df["TBM1"].apply(xep_loai)
df["XL2"] = df["TBM2"].apply(xep_loai)
df["XL3"] = df["TBM3"].apply(xep_loai)

# ===== HIỂN THỊ KẾT QUẢ =====
print(df[[
    "TBM1","XL1",
    "TBM2","XL2",
    "TBM3","XL3"
]].head(10))
```

9. Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp
10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương
pháp Min-Max Normalization

```python
import pandas as pd

# Đọc dữ liệu
df = pd.read_csv("dulieuxettuyendaihoc.csv")

# ===== TẠO TBM1 =====
df["TBM1"] = (
    df["T1"] * 2 +
    df["L1"] +
    df["H1"] +
    df["S1"] +
    df["V1"] * 2 +
    df["X1"] +
    df["D1"] +
    df["N1"]
) / 10

# ===== TẠO TBM2 =====
df["TBM2"] = (
    df["T2"] * 2 +
    df["L2"] +
    df["H2"] +
    df["S2"] +
    df["V2"] * 2 +
    df["X2"] +
    df["D2"] +
    df["N2"]
) / 10

# ===== TẠO TBM3 =====
df["TBM3"] = (
    df["T6"] * 2 +
    df["L6"] +
    df["H6"] +
    df["S6"] +
    df["V6"] * 2 +
    df["X6"] +
    df["D6"] +
    df["N6"]
) / 10

# ===== MIN-MAX NORMALIZATION =====

# US_TBM1
df["US_TBM1"] = (
    (df["TBM1"] - df["TBM1"].min()) /
    (df["TBM1"].max() - df["TBM1"].min())
) * 4

# US_TBM2
df["US_TBM2"] = (
    (df["TBM2"] - df["TBM2"].min()) /
    (df["TBM2"].max() - df["TBM2"].min())
) * 4

# US_TBM3
df["US_TBM3"] = (
    (df["TBM3"] - df["TBM3"].min()) /
    (df["TBM3"].max() - df["TBM3"].min())
) * 4

# ===== HIỂN THỊ KẾT QUẢ =====
print(df[[
    "TBM1", "US_TBM1",
    "TBM2", "US_TBM2",
    "TBM3", "US_TBM3"
]].head(10))
```

10. Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định sinh viên đậu (giá trị “1”) và
rớt ( giá trị “0”) vào các khối dựa trên điểm DH1, DH2 và DH3 như sau

```python
import pandas as pd

# Đọc dữ liệu
df = pd.read_csv("dulieuxettuyendaihoc.csv")

# ===== HÀM XÉT TUYỂN =====
def xet_tuyen(row):

    # Khối A hoặc A1
    if row["KT"] in ["A", "A1"]:

        diem_xt = (
            row["DH1"] * 2 +
            row["DH2"] +
            row["DH3"]
        ) / 4

    # Khối B
    elif row["KT"] == "B":

        diem_xt = (
            row["DH1"] +
            row["DH2"] * 2 +
            row["DH3"]
        ) / 4

    # Các khối còn lại
    else:

        diem_xt = (
            row["DH1"] +
            row["DH2"] +
            row["DH3"]
        ) / 3

    # Kết quả xét tuyển
    if diem_xt >= 5:
        return 1
    else:
        return 0

# ===== TẠO BIẾN KQXT =====
df["KQXT"] = df.apply(xet_tuyen, axis=1)

# ===== HIỂN THỊ KẾT QUẢ =====
print(df[[
    "KT",
    "DH1",
    "DH2",
    "DH3",
    "KQXT"
]].head(20))
```
