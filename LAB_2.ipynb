3. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1,
Q2 và Q3 của DH1 theo KT  

```python
import numpy as np

def q1(x):
    return x.quantile(0.25)

def q2(x):
    return x.quantile(0.50)

def q3(x):
    return x.quantile(0.75)

pivot_table_dh1 = df.pivot_table(values='DH1', index='KT',
                                aggfunc=['count', 'sum', 'mean', 'median', 'min', 'max', 'std', q1, q2, q3])

display(pivot_table_dh1)
```

# Phần 3: Trực quan hóa dữ liệu theo nhóm phân loại

1. Trực quan dữ liệu học sinh nữ trên các nhóm XL1, XL2, XL3 dạng unstacked

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the classification function based on average score
def classify(score):
    if score < 5.0: return 'Y'
    elif score < 6.5: return 'TB'
    elif score < 8.0: return 'K'
    elif score < 9.0: return 'G'
    else: return 'XS'

# Create classification columns for each year
for i in range(1, 4):
    df[f'XL{i}'] = df[f'US_TBM{i}'].apply(classify)

# Filter for female students
female_df = df[df['GT'] == 'F']

# Count classifications for each XL column
xl1_counts = female_df['XL1'].value_counts()
xl2_counts = female_df['XL2'].value_counts()
xl3_counts = female_df['XL3'].value_counts()

# Combine into a single DataFrame for unstacked plotting
# Ensure all categories [Y, TB, K, G, XS] are present in index
categories = ['Y', 'TB', 'K', 'G', 'XS']
xl_summary = pd.DataFrame({
    'XL1': xl1_counts,
    'XL2': xl2_counts,
    'XL3': xl3_counts
}).reindex(categories).fillna(0)

# Plotting
ax = xl_summary.T.plot(kind='bar', figsize=(10, 6), width=0.8)

plt.title('Trực quan hóa xếp loại học sinh nữ (XL1, XL2, XL3)')
plt.xlabel('Nhóm xếp loại theo năm')
plt.ylabel('Số lượng học sinh')
plt.legend(title='Xếp loại')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

2. Trực quan dữ liệu KQXT trên nhóm học sinh có khối thi A, A1, B thuộc khu vực 1, 2

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate KQXT: Pass (1) if average of DH1, DH2, DH3 >= 5.0, else Fail (0)
# (Assuming standard university entrance criteria for this exercise)
df['KQXT'] = df[['DH1', 'DH2', 'DH3']].mean(axis=1).apply(lambda x: 1 if x >= 5.0 else 0)

# Filter the data based on the criteria:
# 1. Blocks: A, A1, B
# 2. Regions: 1, 2
filtered_kqxt = df[
    (df['KT'].isin(['A', 'A1', 'B'])) &
    (df['KV'].isin(['1', '2']))
]

# Count the results
kqxt_counts = filtered_kqxt['KQXT'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(8, 6))
sns.countplot(data=filtered_kqxt, x='KQXT', palette='Set2')

plt.title('Kết quả xét tuyển (KQXT) cho Khối A, A1, B - Khu vực 1, 2')
plt.xlabel('Kết quả (0: Rớt, 1: Đậu)')
plt.ylabel('Số lượng học sinh')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show statistics
print("Thống kê kết quả xét tuyển cho nhóm đối tượng này:")
display(kqxt_counts)

plt.show()
```

2. Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau:

```python
def classify_t1(score):
    if score < 5:
        return 'k'   # kém
    elif score < 7:
        return 'tb'  # trung bình
    elif score < 8:
        return 'kh'  # khá
    else:
        return 'g'   # giỏi

# Tạo biến phanlopt1
df['phanlopt1'] = df['T1'].apply(classify_t1)

# Hiển thị 10 dòng đầu tiên để kiểm tra
display(df[['T1', 'phanlopt1']].head(10))
```

5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanlopt1

```python
import matplotlib.pyplot as plt
import numpy as np

# Thiết lập kích thước biểu đồ
plt.figure(figsize=(16, 8))

# Định nghĩa nhóm và màu sắc
groups = ['k', 'tb', 'kh', 'g']
colors = ['red', 'orange', 'blue', 'green']

# Vẽ drop-line cho từng nhóm
for group, color in zip(groups, colors):
    group_indices = df[df['phanlopt1'] == group].index
    group_values = df[df['phanlopt1'] == group]['T1']

    # Sử dụng stem plot để tạo drop-line
    markerline, stemlines, baseline = plt.stem(group_indices, group_values,
                                                linefmt=color, markerfmt='o',
                                                label=f'Nhóm {group}')

    # Tùy chỉnh kích thước marker và độ dày đường kẻ
    plt.setp(markerline, markersize=4, color=color)
    plt.setp(stemlines, linewidth=0.8, alpha=0.6)

# Thêm các thành phần biểu đồ
plt.title('Biểu đồ Drop-line cho biến T1 phân loại bởi phanlopt1', fontsize=14)
plt.xlabel('Chỉ số học sinh (Index)', fontsize=12)
plt.ylabel('Điểm Toán (T1)', fontsize=12)
plt.legend(title='Phân loại T1')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

# Hiển thị biểu đồ
plt.show()
```

1. Hãy mô tả và khảo sát phân phối cho biến T1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# 1. Mô tả độ tập trung và phân tán
description = df['T1'].describe()
skewness = df['T1'].skew()
kurtosis = df['T1'].kurtosis()
mean_val = df['T1'].mean()
median_val = df['T1'].median()
mode_val = df['T1'].mode()[0]

print("--- Thống kê mô tả T1 ---")
print(description)
print(f"Mean: {mean_val:.2f}, Median: {median_val:.2f}, Mode: {mode_val:.2f}")
print(f"Skewness (Độ lệch): {skewness:.4f}")
print(f"Kurtosis (Độ nhọn): {kurtosis:.4f}")

# 2. Vẽ biểu đồ Box-Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['T1'], color='skyblue', showmeans=True)
plt.title('Biểu đồ Box-Plot của T1')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

# 3. Vẽ biểu đồ Histogram biểu thị hình dáng phân phối
plt.figure(figsize=(10, 6))
sns.histplot(df['T1'], kde=True, color='salmon', bins=10)
plt.axvline(mean_val, color='blue', linestyle='--', label=f'Mean: {mean_val:.2f}')
plt.axvline(median_val, color='green', linestyle='-', label=f'Median: {median_val:.2f}')
plt.title('Biểu đồ Histogram và đường mật độ của T1')
plt.legend()
plt.show()

# 4. Kiểm chứng phân phối chuẩn QQ-Plot
plt.figure(figsize=(8, 6))
stats.probplot(df['T1'], dist="norm", plot=plt)
plt.title('Biểu đồ QQ-Plot của T1')
plt.show()

# Nhận xét
print("\n--- Nhận xét ---")
if abs(skewness) < 0.5:
    skew_label = "Phân phối khá đối xứng"
elif skewness > 0:
    skew_label = "Phân phối lệch phải (tích cực)"
else:
    skew_label = "Phân phối lệch trái (tiêu cực)"

print(f"- Hình dáng: {skew_label}.")
print(f"- Đặc trưng: Với Mean ({mean_val:.2f}) gần bằng Median ({median_val:.2f}), dữ liệu có xu hướng hội tụ về trung tâm.")
print("- QQ-Plot: Các điểm dữ liệu bám khá sát đường thẳng, cho thấy T1 có phân phối xấp xỉ chuẩn.")
```

5. Hãy khảo sát tương quan giữa các biến DH1, DH2,  DH3

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Chọn các biến cần khảo sát
dh_vars = ['DH1', 'DH2', 'DH3']
dh_data = df[dh_vars]

# 2. Tính toán ma trận hiệp phương sai (Covariance Matrix)
cov_matrix = dh_data.cov()
print("--- Ma trận hiệp phương sai (Covariance Matrix) ---")
display(cov_matrix)

# 3. Tính toán ma trận tương quan (Correlation Matrix)
corr_matrix = dh_data.corr()
print("\n--- Ma trận tương quan (Correlation Matrix) ---")
display(corr_matrix)

# 4. Vẽ biểu đồ Heatmap để trực quan hóa ma trận tương quan
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.4f', linewidths=0.5)
plt.title('Heatmap tương quan giữa DH1, DH2 và DH3')
plt.show()

# 5. Vẽ biểu đồ Scatter Matrix (Pairplot)
sns.pairplot(dh_data, diag_kind='kde', plot_kws={'alpha': 0.6, 's': 30, 'edgecolor': 'k'})
plt.suptitle('Biểu đồ Scatter Matrix giữa DH1, DH2, DH3', y=1.02)
plt.show()

# 6. Nhận xét sơ bộ
print("\n--- Nhận xét ---")
print("Dựa vào ma trận tương quan:")
for i in range(len(dh_vars)):
    for j in range(i + 1, len(dh_vars)):
        c = corr_matrix.iloc[i, j]
        strength = 'yếu' if abs(c) < 0.3 else ('trung bình' if abs(c) < 0.7 else 'mạnh')
        direction = 'tích cực (đồng biến)' if c > 0 else 'tiêu cực (nghịch biến)'
        print(f"- {dh_vars[i]} và {dh_vars[j]}: Tương quan {strength} ({c:.4f}), xu hướng {direction}.")
```
